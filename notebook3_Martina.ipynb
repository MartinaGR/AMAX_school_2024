# Fibonacci Sequence and Rabbit Population Modeling

This lab explores the Fibonacci sequence and its application in modeling rabbit populations. We'll start with the traditional Fibonacci model and then introduce a stochastic component to simulate randomness in reproduction rates.

## Part 1: Simple Fibonacci Model

### Modeling Rabbit Populations

The Fibonacci sequence was originally inspired by the growth of a hypothetical rabbit population. Let's use the Fibonacci sequence to model the population of rabbits over several months under the following assumptions:

1. A newly born pair of rabbits (one male, one female) are put in a field.
2. Rabbits are able to mate at the age of one month so that at the end of its second month, a female can produce another pair of rabbits.
3. Rabbits never die, and a mating pair always produces one new pair every month from the second month onwards.

We will model the population of rabbits for one year and analyze the results.

### Implementing the Fibonacci Sequence in Python

```python
def fibonacci(n):
  fib_list = [1,1]
  for i in range(n-1):
    new_value = fib_list[-1] + fib_list[-2]
    fib_list.append(new_value)
  return fib_list
```

```python
fibonacci(13)
```

### Visualization

```python
rabbit_population = fibonacci(13)  # 12 months + the starting month
print("Rabbit pairs after each month:", rabbit_population)

# Plotting the rabbit population growth
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(rabbit_population, marker='o')
plt.title('Rabbit Population Growth Over 12 Months')
plt.xlabel('Month')
plt.ylabel('Number of Rabbit Pairs')
plt.xticks(range(13))
plt.grid(True)
plt.show()
```

## Part 2: Stochastic Fibonacci Model

### Introduction to Stochastic Processes

A stochastic process involves randomness. In the context of rabbit populations, it means that the number of offspring produced can vary according to certain probabilities.

```python
# Probabilities

# 0 offsprings with prob 20%

# 1 par with prob 40%

# 2 par with prob 30%
```

### Implementing a Stochastic Model in Python

```python
import random

def stochastic_fibonacci(n):
  fib_list=[1,1]
  for i in range(2, n+1):
    offsprings_prob = random.random()
    if offsprings_prob < 0.2:
      offsprings_prob = 0
    elif offsprings_prob <0.7:
      offsprings_prob = 1
    else:
      offsprings_prob = 2
    new_value = fib_list[-1] + fib_list[-2]* offsprings_prob
    fib_list.append(new_value)
  return fib_list

print(stochastic_fibonacci(10))
```
